```python
#import pandas and read csv file
import pandas as pd
dataframe=pd.read_csv('https://raw.githubusercontent.com/vigneshk/Admission-Dataset/master/Admission.csv')
```

```python
#print dataframe
dataframe
```

```python
#import pymongo
import pymongo
```

```python
#Create client
client=pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
#Create Database
db1=client['Assignment']
```

```python
#create collection name
collection=db1['Products']
```

```python
#dataframe convert into dict
dataframe.reset_index(inplace=True)
df_dict=dataframe.to_dict('records')
```

```python
#insert the record data to collections
rec=collection.insert_many(df_dict)
```

```python
#print the collections
res=collection.find()
```
